````python
# Order by timestamp
df['Measurement Timestamp'] = pd.to_datetime(df['Measurement Timestamp'])

df = df.set_index('Measurement Timestamp').sort_index()

#output file

df.reset_index().to_csv('output/q3_wrangled_data.csv', index = False)
````

````python
df.isnull().sum()
````

````python
#temporal features of time column
df = df.reset_index()

df_time = pd.DataFrame()
df_time['Measurement Timestamp'] = pd.to_datetime(df['Measurement Timestamp'], format = '%Y-%m-%d %H:%M:%S')
df_time['Hour'] = df_time['Measurement Timestamp'].dt.hour
df_time['Weekday'] = df_time['Measurement Timestamp'].dt.weekday
df_time['Month'] = df_time['Measurement Timestamp'].dt.month
````

````python
#check dataframe

df_time.head()
````

````python
#save to csv

df_time.to_csv('output/q3_temporal_features.csv', index = False)
````

````python
# get date range
date_range_start = df['Measurement Timestamp'].min()
date_range_end = df['Measurement Timestamp'].max()

# save to file
with open('output/q3_datetime_info.txt', 'w') as f:
    f.write("Date Range After Datetime Parsing:\n")
    f.write(f"Start: {date_range_start}\n")
    f.write(f"End: {date_range_end}\n")
````
